# **Step 2: Import Libraries**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import ImageEnhance, Image
```

# **Step 3: Load the image**

```python
image = cv2.imread('image1.png')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()
```

# **STEP 4: GRAYSCALE CONVERSION**

```python
  gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  plt.imshow(gray_image, cmap='gray')
  plt.title('Grayscale Image')
  plt.show()
```

# **STEP 5: GAUSSIAN BLUR**

```python
blurred_image = cv2.GaussianBlur(gray_image, (55, 55), 0)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')
```

# **STEP 6: EDGE DETECTION**

```python
edges = cv2.Canny(blurred_image, 5, 25)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.show()
```

# **7. Histogram Equalization**

```python
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply histogram equalization
    equalized_image = cv2.equalizeHist(gray_image)
    plt.imshow(equalized_image, cmap='gray')
    plt.title("Histogram Equalized")
```

# **8. Illumination-Invariant Color Space (HSV) - Brighten and Darken**

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

hsv_image[..., 2] = cv2.equalizeHist(hsv_image[..., 2])
illumination_invariant_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)
plt.imshow(cv2.cvtColor(illumination_invariant_image, cv2.COLOR_BGR2RGB))
plt.title("Illumination-Invariant (HSV)")

pil_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

# Brighten image
enhancer = ImageEnhance.Brightness(pil_image)
brighter_image = enhancer.enhance(1.5)  # Increase brightness by 1.5 times
brighter_image_cv = cv2.cvtColor(np.array(brighter_image), cv2.COLOR_RGB2BGR)
plt.imshow(cv2.cvtColor(brighter_image_cv, cv2.COLOR_BGR2RGB))
plt.title("Brightened Image")
```

```python
darker_image = enhancer.enhance(0.5)  # Decrease brightness by 50%
darker_image_cv = cv2.cvtColor(np.array(darker_image), cv2.COLOR_RGB2BGR)
plt.imshow(cv2.cvtColor(darker_image_cv, cv2.COLOR_BGR2RGB))
plt.title("Darkened Image")
```

```python

    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # Apply face detection on the equalized image
    faces = face_cascade.detectMultiScale(equalized_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Draw rectangles around detected faces
    detection_image = image.copy()
    for (x, y, w, h) in faces:
        cv2.rectangle(detection_image, (x, y), (x+w, y+h), (255, 0, 0), 2)

    # Display Face Detection Result
    plt.imshow(cv2.cvtColor(detection_image, cv2.COLOR_BGR2RGB))
    plt.title("Face Detection on Equalized Image")

    # Show all images
    plt.tight_layout()
    plt.show()
```
